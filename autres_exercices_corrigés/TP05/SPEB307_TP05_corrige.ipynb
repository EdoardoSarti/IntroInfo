<b><font color='white' style='background-color:orange'>3.1.b</font></b> Pour chaque ligne du fichier `sequences.txt`, affichez le numÃ©ro de ligne et le pourcentage des occurrences des caractÃ¨res `C` et `G` (ensemble).

````python
f = open("sequences.txt")
for line in f:
    count = 0
    norm = 0
    for c in line:
        if c == "C" or c == "G":
            count = count + 1
        norm = norm + 1
    print(count/norm)
f.close()
````
